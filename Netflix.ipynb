## Q.2. In which year highest number of the TV Shows  Movies sere released ? Show with bar graph.

```python
df.head()
```

```python
df.dtypes["Release_Date"]
```

```python
df["Release_Date"]= pd.to_datetime(df["Release_Date"],errors='coerce', format='%B %d, %Y', dayfirst=True) # string ve ayların string bir ifade de olan tarihi pd.to_datetime ile ceviririz
```

```python
df.dtypes
```

```python
tv_show= df[df["Category"]=="TV Show"]
```

```python
tv_show_graph=tv_show["Release_Date"].dt.year.value_counts(ascending=False) #tv show kategorisini yıllara göre topladık.
tv_show_graph
```

```python
plt.figure(figsize=(10,5))
sns.barplot(x=tv_show_graph.index,y=tv_show_graph.values, order=tv_show_graph.index,)
plt.xlabel("Year")
plt.ylabel("Values")
plt.title("TV Show's highest debut years")
plt.show()
```

## Countries with the most TV shows and movies

```python
country_counts=df["Country"].value_counts().reset_index()
country_counts=country_counts.rename(columns={"Country": "Country", "count":"total"})
country_counts["Country"].replace("United States","United States of America", inplace=True)
country_counts=country_counts.sort_values(by="total")
```

```python
import geopandas as gpd
import fiona
```

```python
world_filepath = gpd.datasets.get_path('naturalearth_lowres')
```

```python
world = gpd.read_file(world_filepath)
```

```python
merged_data = world.merge(country_counts, left_on='name', right_on="Country")
```

```python
merged_data
```

```python
merged_data
```

```python
fig, ax = plt.subplots(1, 1, figsize=(15, 10))
merged_data.plot(column='total', cmap='viridis', linewidth=0.8, ax=ax, edgecolor='0.9', legend=True, legend_kwds={'label': "Total Counts"})
ax.set_title('Total Counts by Country')
plt.show()
```

```python
country_counts_sort=country_counts.sort_values(by='total',ascending=False).head(10)
plt.figure(figsize=(10,10))
plt.xticks(rotation=90)
sns.barplot(x=country_counts_sort["Country"],y= country_counts_sort["total"])
plt.show()
```

## Continents with the most TV shows and movies

```python
merged_data2=merged_data.groupby("continent")["total"].sum().reset_index()
```

```python
merged_data2=merged_data2.sort_values(by="total", ascending=False)
```

```python
plt.figure(figsize=(10,10))
plt.xticks(rotation=90)
sns.barplot(x=merged_data2["continent"],y= merged_data2["total"] )
plt.show()
```
